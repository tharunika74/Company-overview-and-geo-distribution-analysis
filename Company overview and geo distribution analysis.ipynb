```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("companies.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python

df.fillna({'state': 'Unknown', 'city': 'Unknown', 'address': 'Unknown', 'company_size': 'Unknown'}, inplace=True)


```

```python

df['zip_code'].replace(0, np.nan, inplace=True)

```

```python

df.isnull().sum()
```

```python
top_countries = df['country'].value_counts().head(15)
```

```python


plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='magma')
plt.title('Top 15 Countries by Company Count')
plt.xlabel('Number of Companies')
plt.ylabel('Country')
plt.show()

```

```python
top_cities = df['city'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='magma')
plt.title('Top 10 Cities with Most Companies')
plt.xlabel('Number of Companies')
plt.ylabel('City')
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['company_size'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('dark'), startangle=140)
plt.title('Company Size Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False, cmap='YlOrBr')
plt.title('Missing Data Heatmap')
plt.show()
```
